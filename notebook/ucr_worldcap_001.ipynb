```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_wc_day37_1 = pd.read_csv(f"../data/wc_day37_1.csv")
df_wc_day38_1 = pd.read_csv(f"../data/wc_day38_1.csv")
df_wc_day38_2 = pd.read_csv(f"../data/wc_day38_2.csv")
df_wc_day39_1 = pd.read_csv(f"../data/wc_day39_1.csv")
df_wc_day39_2 = pd.read_csv(f"../data/wc_day39_2.csv")
df_wc_day40_1 = pd.read_csv(f"../data/wc_day40_1.csv")
df_wc_day40_2 = pd.read_csv(f"../data/wc_day40_2.csv")
df_wc_day41_1 = pd.read_csv(f"../data/wc_day41_1.csv")
df_wc_day41_2 = pd.read_csv(f"../data/wc_day41_2.csv")
```

```python
df = pd.concat([
    df_wc_day37_1,
    df_wc_day38_1,
    df_wc_day38_2,
    df_wc_day39_1,
    df_wc_day39_2,
    df_wc_day40_1,
    df_wc_day40_2,
    df_wc_day41_1,
    df_wc_day41_2,
])
```

```python
df['date'] = pd.to_datetime(df['date'])

df_S = df.groupby(pd.Grouper(key="date", freq="min")).sum().reset_index()

plt.plot(df_S["date"], df_S["count"])
```
